```python
import re
import matplotlib.pyplot as plt
from Bio import SeqIO
import mplcursors


def find_codons(sequence, codon_list):
    """
    Find codon positions in a given nucleotide sequence.
    
    Parameters:
    - sequence (str): The nucleotide sequence (RNA format: U instead of T).
    - codon_list (list): List of codons to search for.
    
    Returns:
    - list: A sorted list of tuples (start position, end position, codon).
    """
    positions = []
    for codon in codon_list:
        for match in re.finditer(codon, str(sequence)):  # Convert sequence to string
            positions.append((match.start(), match.end(), codon))
    return sorted(positions)

# Load the sequence from FASTA file
record = SeqIO.read("Kp326_HBRU285_26.fasta", "fasta")
sequence = record.seq.transcribe()  # Convert DNA to RNA (T → U)

# Find Start and Stop Codons
start_codons = find_codons(sequence, ["AUG"])  # Start codon
stop_codons = find_codons(sequence, ["UAA", "UGA", "UAG"])  # Stop codons

# Print Results
start_codon_list = print(f"Total Start Codons (AUG): {len(start_codons)} at positions: {start_codons}")
stop_codon_list = print(f"Total Stop Codons (UAA, UGA, UAG): {len(stop_codons)} at positions: {stop_codons}")
```
